```python
import matplotlib.pyplot as plt
import numpy as np

# Donnes pour le forward-backward
observations = ['Observation 1', 'Observation 2', 'Observation 3', 'Observation 4']
states = ['Etat 1', 'Etat 2', 'Etat 3']
probabilities = np.array([
    [0.2, 0.3, 0.4, 0.1],  
    [0.3, 0.4, 0.1, 0.2],  
    [0.5, 0.3, 0.5, 0.7]   
])

# Optimal path for Viterbi
optimal_path = [1, 2, 1, 2]  # Optimal path (State indices)

# Initialize figure
fig, axs = plt.subplots(2, 1, figsize=(12, 10))

# Viterbi visualization
axs[0].set_title("Algorithm Viterbi - Recherche de chemin", fontsize=14)
for i, state in enumerate(states):
    axs[0].plot(observations, [i + 1] * len(observations), 'o--', label=state)

# Highlight optimal path
axs[0].plot(observations, [optimal_path[i] for i in range(len(observations))], 'r-', lw=2, label="Chemin optimal")
axs[0].set_yticks(range(1, len(states) + 1))
axs[0].set_yticklabels(states)
axs[0].legend()
axs[0].grid(True)
axs[0].set_ylabel("States")

# Forward-Backward visualization
axs[1].set_title("Algorithme Forward-Backward - Distribution de probabilit", fontsize=14)
width = 0.2  # Bar width for stacked bars
x_obs = np.arange(len(observations))  # Positions for bars

# Stacked bar chart
bottom = np.zeros(len(observations))
for i, state in enumerate(states):
    axs[1].bar(x_obs, probabilities[i], width, bottom=bottom, label=state)
    bottom += probabilities[i]

axs[1].set_xticks(x_obs)
axs[1].set_xticklabels(observations)
axs[1].set_ylabel("Probabilit")
axs[1].legend(title="Etat")
axs[1].grid(axis='y')

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for simulation
states = [5, 10, 20, 50, 100]  # Number of states (N)
sequence_lengths = [100, 500, 1000, 5000]  # Sequence lengths (T)

# Generate execution times (arbitrary values based on complexity O(N^2 * T))
execution_times = np.array([[n**2 * t / 1e6 for t in sequence_lengths] for n in states])

# Plotting
plt.figure(figsize=(12, 8))
for i, n in enumerate(states):
    plt.plot(sequence_lengths, execution_times[i], label=f"{n} States (N)", marker='o')

plt.title("Temps d'execution de l'algorithme de Forward-backward", fontsize=16)
plt.xlabel("Longueur des squences", fontsize=14)
plt.ylabel("temps d'execution (seconds)", fontsize=14)
plt.legend(title="Number of States", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(sequence_lengths)
plt.show()
```
